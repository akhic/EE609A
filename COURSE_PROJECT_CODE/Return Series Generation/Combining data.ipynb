```python
import pandas as pd
import numpy as np
```

```python
stocks = ['ADTB','ASPN','AXBK','BAJA','BJFN','BPCL','BRIT','BRTI','CIPL','COAL','HCLT','HDBK','HDFC','HLL','ICBK','INFY','ITC','JSTL','KTKM','LART','MRTI','PGRD','PIRA','PNBK','REDY','RELI','SAIL','SBI','TCS','WIPR']
```

```python
dt = [pd.read_csv('Historical/'+s+' Historical Data.csv',index_col = 'Date') for s in stocks]
```

```python
valid = '1234567890.' #valid characters for a float
def sanitize(data):
    return float(''.join(filter(lambda char: char in valid, data)))
```

```python
for d in dt:
    d.index = pd.to_datetime(d.index)
    d['Price'] = d['Price'].apply(str)
    d['Price'] = d['Price'].apply(sanitize)
#     d = d.iloc[::-1]
```

```python
for i in range(30):
    dt[i] = dt[i].iloc[::-1]
```

```python
dt[0]
```

```python
for i in range(30):
    dt[i]['log return'] = np.log(dt[i]['Price']/dt[i]['Price'].shift(1))
```

```python
mean_log_ret = np.zeros((30,1))
cov_log_ret = np.zeros((30,30))
```

```python
dt[0]['log return'][1:-300]
```

```python
for i in range(30):
    print(len(dt[i][1:-300]))
```

```python
df = dt[0]
```

```python
for i in range(30):
    df['log return' + str(i)] = dt[i]['log return']
```

```python
df = df.dropna()
```

```python
df = df.drop(columns = ['Price','Open','High','Low','Vol.','Change %','log return'])
```

```python
df
```

```python
df.to_csv('log_returns_full.csv')
```

```python
for i in range(30):
    mean_log_ret[i] = dt[i]['log return'][1:-300].mean()
```

```python
for i in range(30):
    for j in range(30):
        cov_log_ret[i][j] = np.cov(df['log return'+str(i)][:-300],df['log return'+str(j)][:-300])[0,1]
```
